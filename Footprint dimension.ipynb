```python
# @title 📦 FAST EACH CALCULATOR { display-mode: "form" }
# Toca el botón de Play a la izquierda para iniciar

import pandas as pd
import math
from google.colab import files

def run():
    # Entradas rápidas
    L = float(input("L: "))
    W = float(input("W: "))
    H = float(input("H: "))
    N = int(input("N: "))

    # Lógica optimizada
    f = [i for i in range(1, N + 1) if N % i == 0]
    best = (N, 1, 1)
    min_err = float('inf')

    for nx in f:
        rem = N // nx
        for ny in [i for i in range(1, rem + 1) if rem % i == 0]:
            nz = rem // ny
            err = abs((nx/ny) - (L/W)) + abs((ny/nz) - (W/H))
            if err < min_err:
                min_err, best = err, (nx, ny, nz)

    nx, ny, nz = best
    res = pd.DataFrame({"Dim": ["L", "W", "H"], "Box": [L, W, H],
                        "Grid": [nx, ny, nz], "Each": [L/nx, W/ny, H/nz]})

    print(f"\n✅ RESULTADO: {nx}x{ny}x{nz}")
    print(f"📏 MEDIDAS: {L/nx:.3f} x {W/ny:.3f} x {H/nz:.3f}")

    res.to_excel("reporte.xlsx", index=False)
    files.download("reporte.xlsx")

run()
```
